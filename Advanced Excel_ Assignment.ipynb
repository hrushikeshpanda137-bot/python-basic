 1. Explain the difference between Absolute, Relative, and Mixed Cell
Referencing in Excel with examples.

    ans-
    RELATIVE CELL REFERENCE
    The cell reference changes automatically when the formula is copied to another cell.
    EXAMPLE (= A1 + B1),

    ABSOLUTE CELL REFERENCE
    The cell reference remains fixed when the formula is copied. It uses the $ symbol.
    EXAMPLE ($B$1 always remains the same.)

    MIXED CELL REFERENCE
    Either the row or the column is fixed, while the other changes.
    EXAMPLE ($A1 → Column A is fixed; row changes.A$1 → Row 1 is fixed; column changes.)
      

2. What is a Macro in Excel? How does it help in automation?

    A Macro in Excel is a recorded sequence of commands and actions that automates repetitive tasks. Macros are created using the Macro Recorder or written in VBA

How it helps in automation:

    Saves Time: Automates repetitive tasks so you can do them instantly with one click.

    Reduces Errors: Ensures tasks are performed the exact same way every time, eliminating human mistakes.

    Adds Power: Allows you to build custom functions or consolidate data across multiple sheets automatically.

3. What are Text Functions in Excel? Mention any five with examples.

    ans -:  Text functions are formulas used to clean, format, and manipulate alphanumeric data. They are essential for fixing raw data before analysis.

    Five Common Functions:

    TRIM: Removes extra spaces from the beginning, end, or between words.
    Example: =TRIM("  Hello  World  ") → Hello World

    CONCAT: Joins two or more text strings or cell values together.
    Example: =CONCAT("Data", " ", "Analyst") → Data Analyst

    PROPER: Capitalizes the first letter of each word and makes the rest lowercase.
    Example: =PROPER("excel sheet") → Excel Sheet

    LEFT: Extracts a specific number of characters starting from the left side of a cell.
    Example: =LEFT("Python", 2) → Py

    LEN: Counts the total number of characters in a cell (including spaces).
    Example: =LEN("SQL Project") → 11

4. What is the use of Scenario Manager in decision making?

    Scenario Manager is a built-in "What-If Analysis" tool in Excel. It allows you to create, store, and switch between multiple sets of input values (called scenarios) to see how those different inputs affect your final calculated results.
If you are calculating the potential profit of a new business, you can create one scenario where your rent and supplies are expensive, and another where they are cheap. Scenario Manager will instantly calculate the profit for both, helping you decide if the business is financially safe to start.


5. : Define the purpose of VLOOKUP and HLOOKUP. How are they different from XLOOKUP? Which among XLOOKUP and INDEX-MATCH is best while usage?
    VLOOKUP: Searches vertically down the first column of a table to return a value in the same row.
    HLOOKUP: Searches horizontally across the top row of a table to return a value in the same column.
  
How XLOOKUP is Different
    Flexible Direction: VLOOKUP can only search from left to right. XLOOKUP can search in any direction (including right to left).

    Safer Defaults: XLOOKUP defaults to an "Exact Match," whereas older lookups default to "Approximate Match."

    Dynamic: XLOOKUP won't break if you insert or delete columns in your dataset.

XLOOKUP vs. INDEX-MATCH: Which is best?
    
    XLOOKUP is generally the best choice. It is easier to write, less prone to typing errors, and has a built-in feature to handle missing data without needing extra formulas.

    INDEX-MATCH is only best (or necessary) if you are working on an older version of Excel (2019 or older) that does not support the XLOOKUP function.
